```python
class Circle(object):
    
    #class object attributes
    pi = 3.14
    
    def __init__(self,radius=1):
        self.radius = radius
        
    def area(self):
         
            #radius**2 * pi
        
        return (self.radius**2) * Circle.pi
    
    def set_radius(self,new_radius):
        """
        This methods takes in a radius, and resets the current radius of Circle.
        
        """
        self.radius = new_radius
        
    def get_radius(self,):
        return self.radius
```

```python
c = Circle(radius = 10)
```

```python
c.area()
```

```python
c.set_radius(20)
```

```python
c.radius
```

```python
c.get_radius()
```

```python
class Animal(object):
    
    def __init__(self):
        print "Animal Created"
        
    def WhoAmI(self):
        print "Animal"
        
    def eat(self):
        print "Eating"
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print "Dog created"
        
    def WhoAmI(self):
        print "Dog"
        
    def bark(self):
        print "woof!"
        
    
```

```python
d = Dog()
```

```python
d.WhoAmI()
```

```python
d.eat()
```

```python
"""
*Try and except syntax
try :
      you do your operations here...
      
except ExceptionI :
    If there is ExceptionI, then execute this block.
except ExceptionII :
    If there is Exception II, then execute this block.
    .....
    
else:
    If there is no exception then execute this block.

*Finally Syntax
try :
    code block here 
    ....
    due to any exception, this code may be skipped!
    
finally :
    This code block would be always be executed.
"""

```

```python
2 + 's'
```

```python
try:
    2 + 's'
except TypeError: 
    print "there was a type error!"
finally:
    print "Finally this was printed"
```

```python
try:
    f = open('testfile1234','r')  #r/w
    f.write('test write this')
except:
    print 'Error in writting to the file'
    
else:
    print 'File write was a success!'
```

```python
def askint():
    try:
        val = int(raw_input('please enter an integer: '))
        
    except:
        print 'Looks like you did not enter an interger'
        val = int(raw_input("Try again - Please enter an integer: "))
    finally:
        print 'Finally block executed!'
        
    print val
```

```python
#Iterators and generators
```

```python
def gencubes(n):
    for num in range(n):
        yield num**3
```

```python
for x in gencubes(10):
    print (x)
```

```python
def genfibon(n):
    a = 1
    b = 1
    for i in range(n):
        yield a
        t = a
        a = b
        b = t + b    # a,b =b, a+b
        
```

```python
for num in genfibon(10):
    print(num)
```

```python
def simple_gen():
    for x in range(3):
        yield x
```

```python
g = simple_gen()
```

```python
def hello(name = 'Jose'):
    print ('The hello() function has been executed')
    
    def greet():
        return '\t This is inside the greet() function'
    
    def welcome():
        return '\t This is inside the welcome() function'
    
    print(greet())
    print(welcome())
    print('Now we are back inside the Hello() function')
    
```

```python
welcome()
```

```python
def hello(name = 'Jose'):
    def greet():
        return '\t This is inside the greet() function'
    
    def welcome():
        return '\t This is inside the welcome() function'
    
    if name == 'Jose':
        return greet
    else:
        return welcome
```

```python
def new_decorator(func):
    
    def wrap_func():
        print('Code here, before executing the func')
        
        func()
        print('Code here will execute after the func()')
        
    return wrap_func
```

```python
def func_needs_decorator():
    print('This function needs a decorator!')
```

```python
func_needs_decorator = new_decorator(func_needs_decorator)
```

```python
func_needs_decorator()
```

```python
@new_decorator
def func_needs_decorator():
    print('This function needs a decorator!')
```

```python
func_needs_decorator()
```
