Lambda Expressions

Lambda's body is a single expression, not a block of statements
The lambda's body is similar to what we would put in a def body's return
statement. We simplt type the result as an expression instead of explicitly returning it.
Because it is limited to an expression, a lambda is less general that a def 

```python
def square(num):
    result = num**2
    return result
```

```python
square(2)
```

```python
def square(num): return num**2
```

```python
square(2)
```

```python
square2 = lambda num:  num**2
```

```python
square2(10)
```

Nested statements and scope----
      Name references search(at most) four scopes, these are:
 1.Local
 2.Enclosing functions
 3.global
 4.built-in
 
 


```python
#Enclosing function local
```

```python
name = 'This is a global name'
def greet():
    #Enclosing function
    
    name = 'Sammy'
    
    def hello():
        print 'hello ' + name
        
        hello()
greet()
```

```python
print name
```

```python
x =50  #Global statement
def func():
    global x
    print 'This function is now using the global x!'
    print 'Because of global x is: ',x
    x = 2
    print 'Ran func(), changed to global x to ',x 
    

print 'Before calling func(), x is : ', x
func()
print  'Value of x(outside the func()) is: ', x
```
